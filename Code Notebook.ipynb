## Financial Scoring Model

```python
import pandas as pd
import numpy as np

def calculate_financial_health_score(df):
    """
    Calculate a financial health score (0-100) for each family based on various financial indicators.
    
    Parameters:
    df: pandas DataFrame containing family financial data
    
    Returns:
    DataFrame with original data and new financial health score column
    """
    # First, calculate category spending percentages for each family
    category_spending = df.groupby(['Family ID', 'Category'])['Amount'].sum().reset_index()
    total_spending = df.groupby('Family ID')['Amount'].sum().reset_index()
    category_spending = category_spending.merge(total_spending, on='Family ID', suffixes=('', '_total'))
    category_spending['spending_percentage'] = (category_spending['Amount'] / category_spending['Amount_total']) * 100

    # Group by Family ID to get family-level metrics
    family_metrics = df.groupby('Family ID').agg({
        'Income': 'first',
        'Savings': 'first',
        'Monthly Expenses': 'first',
        'Loan Payments': 'first',
        'Credit Card Spending': 'first',
        'Financial Goals Met (%)': 'first'
    }).reset_index()
    
    # 1. Savings-to-Income Ratio (25% weight)
    family_metrics['savings_score'] = (family_metrics['Savings'] / family_metrics['Income'] * 200).clip(0, 100)
    
    # 2. Monthly Expenses to Income Ratio (20% weight)
    family_metrics['expenses_score'] = (100 - (family_metrics['Monthly Expenses'] / family_metrics['Income'] * 100)).clip(0, 100)
    
    # 3. Loan Payments to Income Ratio (15% weight)
    family_metrics['loan_score'] = (100 - (family_metrics['Loan Payments'] / family_metrics['Income'] * 250)).clip(0, 100)
    
    # 4. Credit Card Spending to Income Ratio (15% weight)
    family_metrics['credit_score'] = (100 - (family_metrics['Credit Card Spending'] / family_metrics['Income'] * 300)).clip(0, 100)
    
    # 5. Spending Pattern Score (10% weight)
    category_weights = {
        'Travel': 0.6,
        'Entertainment': 0.7,
        'Dining': 0.8,
        'Shopping': 0.8,
        'Groceries': 1.0,
        'Healthcare': 1.0,
        'Education': 1.0,
        'Utilities': 1.0
    }
    
    # Calculate weighted spending score for each family
    category_spending['category_weight'] = category_spending['Category'].map(lambda x: category_weights.get(x, 0.9))
    category_spending['weighted_score'] = category_spending['spending_percentage'] * category_spending['category_weight']
    spending_pattern_scores = category_spending.groupby('Family ID')['weighted_score'].mean().reset_index()
    spending_pattern_scores['weighted_score'] = spending_pattern_scores['weighted_score'].clip(0, 100)
    
    # Merge spending pattern scores with family metrics
    family_metrics = family_metrics.merge(spending_pattern_scores, on='Family ID')
    family_metrics['spending_pattern_score'] = family_metrics['weighted_score']
    
    # 6. Financial Goals Score (15% weight)
    family_metrics['goals_score'] = family_metrics['Financial Goals Met (%)']
    
    # Calculate final weighted score
    family_metrics['financial_health_score'] = (
        family_metrics['savings_score'] * 0.25 +
        family_metrics['expenses_score'] * 0.20 +
        family_metrics['loan_score'] * 0.15 +
        family_metrics['credit_score'] * 0.15 +
        family_metrics['spending_pattern_score'] * 0.10 +
        family_metrics['goals_score'] * 0.15
    ).round(2)
    
    # Map scores back to original dataframe
    score_mapping = family_metrics[['Family ID', 'financial_health_score']].set_index('Family ID')['financial_health_score']
    df['Financial Health Score'] = df['Family ID'].map(score_mapping)
    
    return df

# Add health score column to the dataframe
df = calculate_financial_health_score(df)
```

## Justification

1. Savings-to-Income Ratio (25% weight)
- Highest weight (25%) because savings is crucial for:
  - Emergency funds
  - Long-term financial security
  - Future investments and opportunities
- Scoring logic: (Savings/Income Ã— 200), capped at 100
- Target: Ideal savings should be at least 50% of income
- Rationale: Higher savings relative to income indicates better financial discipline and security

2. Monthly Expenses to Income Ratio (20% weight)
- Second-highest weight due to its importance in day-to-day financial management
- Scoring logic: 100 - (Monthly Expenses/Income Ã— 100), capped at 100
- Target: Expenses should not exceed 70% of income
- Rationale: 
  - Lower expenses-to-income ratio indicates better budgeting
  - Allows more room for savings and investments
  - Provides buffer for unexpected expenses

3. Loan Payments to Income Ratio (15% weight)
- Scoring logic: 100 - (Loan Payments/Income Ã— 250), capped at 100
- Target: Loan payments should be below 40% of income
- Rationale:
  - High debt burden increases financial risk
  - Impacts ability to save and invest
  - Affects credit worthiness
  - Factor in debt-to-income ratio standards used by lenders

4. Credit Card Spending to Income Ratio (15% weight)
- Scoring logic: 100 - (Credit Card Spending/Income Ã— 300), capped at 100
- Target: Credit card spending should be below 30% of income
- Rationale:
  - High credit card usage often indicates cash flow problems
  - Risk of high-interest debt accumulation
  - Impacts credit utilization ratio
  - Important for short-term financial health

5. Discretionary Spending Pattern Score (10% weight)
- Lower weight as some discretionary spending is normal and necessary
- Category weights:
  - Essential categories (1.0): Groceries, Healthcare, Education, Utilities
  - Moderate impact (0.8): Dining, Shopping
  - Higher impact (0.6-0.7): Travel, Entertainment
- Rationale:
  - Balances necessity vs. luxury spending
  - Rewards essential spending
  - Accounts for quality of life while maintaining financial responsibility

6. Financial Goals Met (15% weight)
- Directly uses the provided Financial Goals Met percentage
- Rationale:
  - Indicates financial discipline and planning ability
  - Measures commitment to financial improvement
  - Reflects ability to set and achieve financial targets
  - Shows long-term financial management capability

Additional Considerations in the Scoring System:

1. Scale Normalization:
- All scores are normalized to 0-100 scale
- Makes components comparable and easy to understand
- Allows for consistent weighting

2. Weight Distribution:
- Core financial health indicators (savings, expenses) = 45%
- Debt management (loans, credit cards) = 30%
- Behavioral factors (spending patterns, goals) = 25%
- This distribution prioritizes fundamental financial stability while considering behavioral aspects

3. Scoring Thresholds:
- Based on common financial planning guidelines
- Aligned with industry standards for debt-to-income ratios
- Considers realistic achievable targets for average families

4. Score Interpretation:
- 90-100: Excellent financial health
- 80-89: Good financial management
- 70-79: Satisfactory financial condition
- 60-69: Needs improvement
- Below 60: Requires significant attention

This scoring system provides a balanced assessment of financial health by:
- Emphasizing long-term financial security (savings)
- Considering day-to-day financial management (expenses)
- Including debt management practices
- Accounting for spending behavior
- Incorporating goal achievement
- Using industry-standard financial ratios and benchmarks

## Linear Regression

```python
x=df.drop(columns=['Financial Health Score'])
```

```python
y=df['Financial Health Score']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoders
category_encoder = LabelEncoder()
day_encoder = LabelEncoder()

# Fit encoders on the training data
X_train['Category'] = category_encoder.fit_transform(X_train['Category'])
X_train['Day of Week'] = day_encoder.fit_transform(X_train['Day of Week'])

# Transform the test data using the fitted encoders
X_test['Category'] = category_encoder.transform(X_test['Category'])
X_test['Day of Week'] = day_encoder.transform(X_test['Day of Week'])

# If you want to see the mappings
category_mapping = dict(zip(category_encoder.classes_, category_encoder.transform(category_encoder.classes_)))
day_mapping = dict(zip(day_encoder.classes_, day_encoder.transform(day_encoder.classes_)))

print("Category Mapping:", category_mapping)
print("Day of Week Mapping:", day_mapping)
```

### Multivariate Normality

```python
import scipy.stats as stats

# Fit a simple linear regression model
from sklearn.linear_model import LinearRegression


model = LinearRegression().fit(X_train, y_train)
residuals = y_test - model.predict(X_test)

# Histogram
sns.histplot(residuals, kde=True)
plt.title("Histogram of Residuals")
plt.show()

# Q-Q Plot
stats.probplot(residuals, dist="norm", plot=plt)
plt.show()
```

### Homoscedasticity

```python
# Residuals vs Predicted
y_pred = model.predict(X_test)
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuals vs Predicted Values")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

The variation of residuals is not evenly distributed, indicating heteroscedasticity. This violates one of the key assumptions of linear regression.

### Independence of Errors

```python
from statsmodels.stats.stattools import durbin_watson

# Durbin-Watson Test
dw_stat = durbin_watson(residuals)
print("Durbin-Watson statistic:", dw_stat)
```

This value is extremely close to 2, which suggests that there is no significant autocorrelation in the residuals.
This supports the assumption of independent residuals, a key assumption in linear regression

```python
from sklearn.metrics import mean_squared_error, r2_score

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Number of samples (rows) and features (columns) in the test set
n = X_test.shape[0]  # Number of observations
p = X_test.shape[1]  # Number of predictors

# Compute R2 score
r2 = r2_score(y_test, y_pred)

# Compute Adjusted R2 score
adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

# Evaluate the model
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
print("Adjusted R2 Score:", adjusted_r2)
```

```python
import pickle

with open('Linear_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
